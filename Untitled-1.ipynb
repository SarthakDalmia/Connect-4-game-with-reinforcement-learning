```python
class Outcome(enum.Enum):
   Win = 1
   Loss = -1
   Draw = 0
   NonTerminal = 2

class Node:
    '''Representation of a node in Tree'''
    '''Turn nodes to dict btw borad and node to make same nodes same'''
    '''Does re running of algorithm stsart from root or current board while picking action??????'''
    '''Do we discard the tree aftere picking an action or grow for other actions??'''
    
    def __init__(self, board, parent):
        self.board = board
        self.parent = parent
        self.children = None

        self.rewards = 0
        self.plays = 0

        self.outcome = self.__calculateOutcome()

    def registerOutcome(self, outcome):
        self.plays+=1
        if outcome == Outcome.Win:
            self.rewards+=1
        elif outcome == Outcome.Loss:
            self.rewards+=-1

    def calculateOutcomeHorizontally(self, board):
        for row in range(board.shape[0]):
            consecutiveNegOnes = 0
            consecutiveOnes = 0
            for col in range(board.shape[1]):
                if board[row][col]==-1:
                    consecutiveNegOnes+=1
                    consecutiveOnes=0
                elif board[row][col]==1:
                    consecutiveNegOnes=0
                    consecutiveOnes+=1
                else:
                    consecutiveNegOnes=0
                    consecutiveOnes=0
                
                if consecutiveNegOnes >= 4:
                    return Outcome.Loss
                if consecutiveOnes >= 4:
                    return Outcome.Win

    def checkDiagonal(self, startRow, startCol, rowUpdate, colUpdate):
        consecutiveNegOnes = 0
        consecutiveOnes = 0

        row = startRow
        col = startCol
        while row >= 0 and col >=0 and row < self.board.shape[0] and col < self.board.shape[1]:
            if self.board[row][col]==-1:
                consecutiveNegOnes+=1
                consecutiveOnes=0
            elif self.board[row][col]==1:
                consecutiveNegOnes=0
                consecutiveOnes+=1
            else:
                consecutiveNegOnes=0
                consecutiveOnes=0
            if consecutiveNegOnes >= 4:
                return Outcome.Loss
            if consecutiveOnes >= 4:
                return Outcome.Win
            
            row+=rowUpdate
            col+=colUpdate

    def __calculateOutcome(self):
        # Calculate state if not win or loss
        outcome = Outcome.Draw
        for col in range(5):
            if self.board[0][col]==0:
                outcome = Outcome.NonTerminal

        newOutcome = self.calculateOutcomeHorizontally(self.board)
        if newOutcome!=None:
            return newOutcome

        newOutcome = self.calculateOutcomeHorizontally(self.board.T)
        if newOutcome!=None:
            return newOutcome
        
        for diags in range(10):
            row = max(0, 5-diags)
            col = max(0,diags-5)
            tempOutcome = self.checkDiagonal(row, col, 1, 1)
            if tempOutcome != None:
                return tempOutcome

        for diags in range(10):
            row = min(5, diags)
            col = max(0,diags-5)
            tempOutcome = self.checkDiagonal(row, col, -1, 1)
            if tempOutcome != None:
                return tempOutcome

        return outcome

    
    def getOutcome(self):
        return self.outcome

    def computeChildren(self):
        self.children = {}
        for move in range(5):
            child = self.createChild(move)
            if child !=None:
                self.children[move] = child
            
    def createChild(self, move):
        last_zero_index = None
        for row in range(6):
            if self.board[row][move] != 0:
                break
            last_zero_index = row
                
        if last_zero_index != None:
            child_board = self.board.copy()
            child_board[last_zero_index][move] = 1
            child_board = -1*child_board
            return Node(child_board, self)

    
    '''(Instead of total number of wins, you can use total rewards to estimate
    the value of a state in MCTS'''
    def uct_value(self, uct_constant):
        if self.parent != None:
            if self.plays == 0:
                return math.inf
            return self.rewards/self.plays + (uct_constant * math.sqrt(math.log(self.parent.plays)/self.plays))
    
    def select_child(self, uct_constant):
        best_uct_value = max([child.uct_value(uct_constant) for child in self.children.values()])
        return random.choice([child for child in self.children.values() if child.uct_value(uct_constant)==best_uct_value])
```

```python
  
class MonteCarloTreeSearchAgent:
    def __init__(self, numSims, uctConst = 1):
        self.root = Node(np.zeros((6,5)).astype(int), None)
        self.numSims = numSims
        self.uctConst = uctConst

    def reset(self):
        self.root = Node(np.zeros((6,5)).astype(int), None)

    def registerMove(self, move):
        if self.root.children == None:
            self.root.computeChildren()
        self.root = self.root.children[move]
        self.root.parent = None

    def getMove(self):
        for sim in range(self.numSims):
            currNode = self.selection()
            
            if currNode.getOutcome() == Outcome.NonTerminal:
                currNode = self.expansion(currNode=currNode)
                
            outcome = self.simulation(currNode=currNode)
        
            self.backprop(currNode=currNode, outcome=outcome)

        # Calculate best move
        bestMove = None
        maxPlays = -1
        for move in self.root.children:
            if self.root.children[move].plays > maxPlays:
                maxPlays = self.root.children[move].plays
                bestMove = move
        
        self.registerMove(bestMove)
        return bestMove

    def selection(self):
        currNode = self.root
        while currNode.children!=None and currNode.getOutcome()==Outcome.NonTerminal:
            currNode = currNode.select_child(self.uctConst)
        return currNode
    
    def expansion(self, currNode):
        currNode.computeChildren()
        return currNode.select_child(self.uctConst)
    
    def simulation(self, currNode):
        if currNode.getOutcome() == Outcome.NonTerminal:
            actions = [i for i in range(5) if currNode.board[0][i]==0]
            nextBoard = currNode.createChild(random.choice(actions))
            nextBoard.parent = None

            outcome = self.simulation(nextBoard)
            if outcome == Outcome.Win:
                return Outcome.Loss
            elif outcome == Outcome.Loss:
                return Outcome.Win
            else:
                return Outcome.Draw
        else:
            return currNode.getOutcome()

    def backprop(self, currNode, outcome):
        while currNode!=None:
            currNode.registerOutcome(outcome)
            if outcome == Outcome.Win:
                outcome =  Outcome.Loss
            elif outcome == Outcome.Loss:
                outcome = Outcome.Win
            currNode = currNode.parent
```

```python
'''Make board and node seprate classes'''
class Connect4:
    def __init__(self, agent1, agent2):
        self.wins = [0, 0]
        self.agents = [agent1, agent2]
    
    def play(self, firstAgent):
        self.agents[0].reset()
        self.agents[1].reset()
        board = Node(np.zeros((6,5)).astype(int), None)
        currAgent = firstAgent
        while True:
            move = self.agents[currAgent].getMove()
            self.agents[1-currAgent].registerMove(move)

            board = board.createChild(move)
            board.parent = None
            
            if board.getOutcome() != Outcome.NonTerminal:
                print("Agent",currAgent,board.getOutcome())
                if board.getOutcome() == Outcome.Win:
                    self.wins[currAgent]+=1
                else:
                    self.wins[1-currAgent]+=1
                break
            currAgent = 1-currAgent
    

game = Connect4(MonteCarloTreeSearchAgent(40, 1.4), MonteCarloTreeSearchAgent(200, 1.4))
for i in range(5):
    game.play(0)

print(game.wins)

game = Connect4(MonteCarloTreeSearchAgent(40, 1.4), MonteCarloTreeSearchAgent(200, 1.4))
for i in range(5):
    game.play(1)

print(game.wins)
```
